```python
# See the Sales of Category
plt.figure(figsize=(10,5))
with sns.axes_style('darkgrid'):
  sns.barplot(data=df,x='Category',y='Sales',color="#FA8FA1",width=0.5)
  plt.title("Sales Vs Category ")
  plt.show()
```

```python

x=df['Region'].value_counts()
y=df['Region'].value_counts().index
ex=[0,0,0,0.1]
plt.pie(x,labels=y,shadow=True,explode=ex,radius=1.2,colors=['#04FBE7','#30FC8C','#FB0429','#FCBF30'],autopct='%0.2f%%')
plt.title('Region',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(60,45))
with sns.axes_style('darkgrid'):
  sns.barplot(data=df,y='Sales',x='State',palette="viridis")
  plt.title("Sales Vs State",fontsize=75)
  plt.xticks(rotation=60,fontsize=40)
  plt.xlabel('State',fontsize=40)
  plt.ylabel('Sales',fontsize=40)
  plt.tight_layout()
  plt.show()
```

```python
plt.figure(figsize=(10,5))
with sns.axes_style('whitegrid'):
  sns.kdeplot( data=df,
    x='Sales',
    fill=False)
  plt.xlim(-1000,5000)
  plt.title('Sales',fontsize=15)

  plt.show()
```

```python
plt.figure(figsize=(10,5))
with sns.axes_style('whitegrid'):
  sns.kdeplot( data=df,
    x='Sales',
    hue='Segment',
    fill=False)
  plt.xlim(-1000,5000)
  plt.title('Sales VS Segement',fontsize=15)
  plt.show()
```

```python
plt.figure(figsize=(10,5))
with sns.axes_style('darkgrid'):
  sns.barplot(data=df,x='Category',y='Sales',palette="viridis",width=0.6,hue='Segment')
  plt.title("Sales Vs Category",fontsize=15)
  plt.show()
```

```python
# See the Sales of Category
plt.figure(figsize=(10,5))
with sns.axes_style('darkgrid'):
  sns.barplot(data=df,x='Category',y='Sales',palette='rocket',hue='Ship Mode')
  plt.title("Sales Vs Category",fontsize=15)
  plt.show()
```

```python
ex=[0,0,0.1]

plt.pie(df['Segment'].value_counts(),labels=df['Segment'].value_counts().index,colors=['#F5C894','#9DFF5C','#FF8AB9'],autopct='%0.2f%%',
shadow=True,explode=ex,radius=1.2,startangle=20)

plt.title('Segment',fontsize=15)
plt.show()
```

```python
# See the Sales of Sub Category
plt.figure(figsize=(10,5))
with sns.axes_style('darkgrid'):
  sns.barplot(data=df,x='Sub-Category',y='Sales',color='#EFFF2E',width=1)
  plt.title("Sales Vs Sub-Category",fontsize=15)

  plt.xticks(rotation=63)
  plt.show()
```

```python
sns.relplot(data=df,x='Sales',y='Profit',kind='scatter',color='r')
plt.title('Sales VS Profit',fontsize=15)
plt.grid()
plt.show()
```

```python
# See the Sales of sub-Category
plt.figure(figsize=(50,7))
with sns.axes_style('darkgrid'):
  sns.catplot(data=df,x='Sub-Category',y='Sales',palette="Set2",hue='Category',kind='bar')
  plt.title("Sales Vs Sub-Category")
  plt.xticks(rotation=55)
  plt.show()
```
