```python
fil=df['Gender'].value_counts()
fil.index
```

```python


plt.pie(fil,labels=fil.index,autopct='%1.2f%%')
plt.title('Gender Overview')
plt.show()
```

```python
plt.bar(df['Pclass'],df['Age'],color='#BFE070',width=0.4)
plt.xlabel('Public Class')
plt.ylabel('Age')
plt.title('Pclass VS Age')

plt.show()
```

```python
plt.pie(df['Pclass'].value_counts(),labels = ['First Class', 'Second Class', 'Third Class'],autopct='%1.2f%%')
plt.title('Pclass Overview')
plt.show()
```

```python
plt.bar(df['Embarked'],df['Age'],color='#41E8AC',width=0.5)
plt.xlabel('Embarked')
plt.ylabel('Age')
plt.title('Embarked VS Age')
plt.show()
```

```python
plt.hist(df['Fare'],bins=[10,20,30,40,50,60,70,80],histtype='barstacked',edgecolor='red',color='#44EB9A',cumulative=-1)

plt.title('Ticket Price Each Passenger Paid')
plt.show()
```

```python
plt.bar(df['SibSp'],df['Age'],color='#9057DE')
plt.xlabel('Sibling & Spouse')
plt.ylabel('Age')
plt.title('Sibling & Spouse VS Age')



plt.show()
```
